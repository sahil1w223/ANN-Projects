```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
import pickle
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import os

print(os.path.getsize('label_encoder_gender.pkl'))
```

```python
model = load_model('model.h5')


with open('one_hot_encoder_geography.pkl', 'rb') as file:
    one_hot_encoder_geography = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)

```

```python
input_data = {
    'CreditScore': 600,
    'Geography':'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000,
}
```

```python
input_data_df = pd.DataFrame([input_data])
gender_encoder = label_encoder_gender.transform([[input_data['Gender']]])
input_data_df['Gender'] = pd.DataFrame(gender_encoder,columns = ['Gender'])
geo_encoder = one_hot_encoder_geography.transform([[input_data['Geography']]]).toarray()
geo_encoder_df = pd.DataFrame(geo_encoder,columns=one_hot_encoder_geography.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
df = pd.concat([input_data_df.drop('Geography',axis=1), geo_encoder_df],axis = 1)
df
```

```python
scaler_df = scaler.transform(df)
```

```python
y_pred = model.predict(scaler_df)
y_pred
```

```python
pred_pro = y_pred[0][0]
```

```python
if  pred_pro >= 0.5:
    print("The Customer is likely to churn.")
else:
    print('TheCustomer is not likely to churn.')
```
